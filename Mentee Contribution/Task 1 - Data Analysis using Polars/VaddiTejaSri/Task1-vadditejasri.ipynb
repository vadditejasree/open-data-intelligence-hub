```python
import numpy as np
```

```python
#1. Create a NumPy array containing numbers from 1 to 10.
a=np.arange(1,11)#starts at 1 and stops before 11, producing the integers 1 through 10.
print(a)
```

```python
#2. Create an array of 5 zeros.
a=np.zeros(5)#without any datatype declaration it stores float value
print(a)
```

```python
#3. Create an array of 5 ones.
a=np.ones(5,dtype=int)
print(a)
```

```python
#4. Create an array containing only even numbers from 2 to 20.
a=np.arange(2,21,2)#2 is step size
print(a)
```

```python
#5.shape of the matrix
a= np.array([[2,3], [4,6]])
print(a.shape)
```

```python
#6.size of an array
a=np.array([1,2,3,4,5])
print(a.size)
```

```python
#7.find 1st,3rd, and last value in the array
arr=np.array([1,2,3,4,5])
print(arr[0])# First
print(arr[-1])# Last
print(arr[2])# Third
```

```python
#8.replace the index of 3 or a[2]
a=np.array([1,2,3,4,5])
a[2]=100
print(a)
```

```python
#9.create 3*3 matrix with zeros
a=np.zeros((3,3))
print(a)
#10.create 4*4 matrix with ones
b=np.ones((4, 4))
print(b)
```

```python
#11.create an identity 5*5 matrix
a=np.eye(5,dtype=int)
print(a)
```

```python
#12. find 10 random between 1 to 100
a=np.random.randint(1,101,10)
print(a)
```

```python
#13. min and max of an arr
arr=np.array([10,25,5,80,45])
print(arr.max())
print(arr.min())
```

```python
#14.sum of array
z=np.array([1,2,3,4,5])
print(z.sum())
#15.mean of array
q=np.array([10,20,40,50])
print(q.mean())
```

```python
#16.sorting of array
ar=np.array([5,2,8,1,9])
print(np.sort(ar))
#17.reverse
arr=np.array([1,2,3,4,5])
print(arr[::-1])
```

```python
#18.add values greater than 50 in array
c=np.array([10,60,30,80,45,90])
print(c[c>50])
```

```python
#19.cout the value of nonzero in arr
arr=np.array([1,2,2,3,2,4])
print(np.count_nonzero(arr==2))
```

```python
#20.reshape the arr
arr=np.arange(1,13)
print(arr.reshape(3,4))
#21.flatten the arr
arr=np.array([[1,2],[3,4]])
print(arr.flatten())
```

```python
#22.add 2 matrix
a=np.array([1,2,3])
b=np.array([4,5,6])
print(a+b)
#23.subtract 2 matrix
a=np.array([1,2,3])
b=np.array([4,5,6])
print(a-b)
#24.multiply 2 matrix
a=np.array([1,2,3])
b=np.array([4,5,6])
print(a*b)
#25.divide 2 matrix
a=np.array([1,2,3])
b=np.array([4,5,6])
print(a/b)
```

```python
#26.row-wise sum of a matrix
a=np.array([[1,2,3],[4,5,6]])
print(a.sum(axis=1))
#27.column-wise sum of a matrix
a=np.array([[1,2,3],[4,5,6]])
print(a.sum(axis=0))
```

```python
#28.Extract the first row from a matrix
arr = np.array([[1,2,3],[4,5,6]])
print(arr[0])
#29.Extract the last row from a matrix
arr = np.array([[1,2,3],[4,5,6]])
print(arr[-1])
```
