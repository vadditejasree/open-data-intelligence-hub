DATA CLEANING

Clean Column Names

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ","_")
)

df.columns
```

Remove Duplicates

```python
df = df.drop_duplicates()

print("Duplicates removed")
```

Convert Date Columns

```python
df["order_date"] = pd.to_datetime(
    df["order_date"],
    errors="coerce"
)


df["ship_date"] = pd.to_datetime(
    df["ship_date"],
    errors="coerce"
)


df.dtypes
```

FILTERING

High Sales Products

```python
high_sales = df[df["sales"] > 500]

high_sales.head()
```

Technology Products

```python
technology = df[
    df["category"]=="Technology"
]

technology.head()
```

Consumer Segment

```python
consumer = df[
    df["segment"]=="Consumer"
]

consumer.head()
```

SORTING

Top 10 Sales Records

```python
top_sales = df.sort_values(
    by="sales",
    ascending=False
)

top_sales.head(10)
```

Category Sales Summary

```python
category_summary = df.groupby("category").agg(

    total_sales=("sales","sum"),

    average_sales=("sales","mean"),

    number_of_orders=("order_id","count")

).reset_index()


category_summary
```

Region Sales Summary

```python
region_summary = df.groupby("region").agg(

    total_sales=("sales","sum"),

    orders=("order_id","count")

).reset_index()


region_summary
```

Sub Category Analysis

```python
subcategory_summary = df.groupby("sub-category")["sales"].sum()


subcategory_summary.sort_values(
    ascending=False
).head(10)
```

FEATURE ENGINEERING

Create Year Feature

```python
df["year"] = df["order_date"].dt.year

df.head()
```

Create Month Feature

```python
df["month"] = df["order_date"].dt.month

df.head()
```

Create Sales Category

```python
def sales_category(x):

    if x >= 500:
        return "High"

    elif x >=100:
        return "Medium"

    else:
        return "Low"


df["sales_category"] = df["sales"].apply(
    sales_category
)


df.head()
```

VISUALIZATION

Sales by Category

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=category_summary,
    x="category",
    y="total_sales"
)


plt.title("Total Sales by Category")

plt.xticks(rotation=45)

plt.show()
```

Sales Distribution

```python
plt.figure(figsize=(8,5))

plt.hist(
    df["sales"],
    bins=30
)


plt.title("Sales Distribution")

plt.xlabel("Sales")

plt.ylabel("Frequency")

plt.show()
```

Region Sales

```python
plt.figure(figsize=(8,5))


sns.barplot(
    data=region_summary,
    x="region",
    y="total_sales"
)


plt.title("Sales by Region")

plt.show()
```

Category Boxplot

```python
plt.figure(figsize=(8,5))


sns.boxplot(
    data=df,
    x="category",
    y="sales"
)


plt.xticks(rotation=45)

plt.title("Sales Distribution by Category")

plt.show()
```

Numerical Columns

```python
numeric_df = df.select_dtypes(
    include="number"
)


numeric_df.head()
```

Correlation Matrix

Heatmap
